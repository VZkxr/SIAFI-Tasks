```python
# Librerías
import seaborn as sns
import pandas as pd
```

```python
# Carga de los datos
data = pd.read_csv('data.csv')

# Primer sondeo
data.info()
```

```python
data.var()
```

```python
# Filtrado de columnas numéricas donde pueda operar el método de var()
num_data = data.select_dtypes(include=['float64', 'int64'])

var_data = num_data.var()
print(var_data)
```

```python
# Para datos categóricos (tipo objeto):
cat_data = data.select_dtypes(include='O').columns

for c in cat_data:
    print('\Columna: ',c)
    print(data[c].value_counts())
```

```python
n_columns = data.select_dtypes(include='number').columns

matr = data[n_columns].corr()

sns.heatmap(matr, annot=True)
```
